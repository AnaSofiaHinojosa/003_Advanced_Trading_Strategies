```python
import mlflow

client = mlflow.MlflowClient()
for rm in client.search_registered_models():
    print(rm.name)
```

```python
for model_name in ["CNNtrading", "MLPtrading"]:
    print(f"\nModel: {model_name}")
    for mv in client.get_latest_versions(model_name):
        print(f"Version: {mv.version}, Stage: {mv.current_stage}, Run ID: {mv.run_id}")
```

```python
import mlflow

def load_model_by_run_id(run_id: str):
    model_uri = f"runs:/{run_id}/model"
    model = mlflow.tensorflow.load_model(model_uri=model_uri)
    return model

# Example
cnn_model = load_model_by_run_id("065e198155134b87a2258b1ec7715cd9")
mlp_model = load_model_by_run_id("041c1c94d19c45838e2a49471f94fd48")
```
